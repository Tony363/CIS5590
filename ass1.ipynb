```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
stop_words
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from collections import Counter
import string


# Function to clean and stem text
def clean_and_stem(
    text:str,
    ps:PorterStemmer,
    stop_words:set,
)->list:

    # Remove punctuation and convert to lowercase
    text = text.translate(str.maketrans('', '', string.punctuation)).lower()
    # Tokenize the text
    tokens = word_tokenize(text)
    # Stem each word in the text
    return ' '.join([ps.stem(word) for word in tokens if not word.lower() in stop_words])

ps = PorterStemmer() 
b1['stemmed'] = b1['text'].apply(lambda text:clean_and_stem(text,ps,stop_words))
stemmed_word_freq = b1.stemmed.str.split(expand=True).stack().value_counts().sort_values(ascending=False).head(10)
stemmed_word_freq
```

```python
plt.bar(stemmed_word_freq.index,stemmed_word_freq.values)
plt.show()
```
