```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
stop_words
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from collections import Counter
import string


# Function to clean and stem text
def clean_and_stem(
    text:str,
    ps:PorterStemmer,
    stop_words:set,
)->list:

    # Remove punctuation and convert to lowercase
    text = text.translate(str.maketrans('', '', string.punctuation)).lower()
    # Tokenize the text
    tokens = word_tokenize(text)
    # Stem each word in the text
    return ' '.join([ps.stem(word) for word in tokens if not word.lower() in stop_words])

ps = PorterStemmer() 
b1['stemmed'] = b1['text'].apply(lambda text:clean_and_stem(text,ps,stop_words))
stemmed_word_freq = b1.stemmed.str.split(expand=True).stack().value_counts().sort_values(ascending=False).head(10)
stemmed_word_freq
```

```python
plt.bar(stemmed_word_freq.index,stemmed_word_freq.values)
plt.show()
```

#### Rating 1

```python
rating1 = df[df.stars == 1].sample(n=50, random_state=42, replace=False)
rating1.head()
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
import re

# Ensure necessary NLTK downloads
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')


# Function to extract noun-adjective pairs
def extract_noun_adj_pairs(sentences):
    pairs = []
    # sentences = sent_tokenize(text)

    for sentence in sentences:
        # Remove non-alphabetic characters for cleaner POS tagging
        sentence = re.sub('[^a-zA-Z\s]', '', sentence)
        tagged = pos_tag(word_tokenize(sentence))

        for i, (word, tag) in enumerate(tagged[:-1]):
            if tag.startswith('JJ') and tagged[i + 1][1].startswith('NN'):
                pairs.append((tagged[i + 1][0], word))
    return pairs

# Extract pairs
noun_adj_pairs = extract_noun_adj_pairs(rating1.text.tolist())

# Count and get the most common pairs
most_common_pairs = Counter(noun_adj_pairs).most_common(10)

print("Top-10 most frequent noun-adjective pairs:")
for pair, freq in most_common_pairs:
    print(f"{pair}: {freq}")
```

```python
from wordcloud import WordCloud, STOPWORDS


# instantiate a word cloud object
stopwords = set(STOPWORDS) # remove any redundant stopwords
my_cloud = WordCloud( # Create wordcloud by excluding words in stopwords
    background_color='white',
    max_words=2000,
    stopwords=stopwords, 
)
# Use this for fine-tuning :  words that you dont want to appear in wordcloud
stopwords.add('nonsense-s#it') 
# generate the word cloud
my_cloud.generate(three_bus[0])

# display the word cloud
fig = plt.figure(figsize=(14, 18))
plt.imshow(my_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
